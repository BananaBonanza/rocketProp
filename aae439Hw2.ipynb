## Appendix A: Python code for 3.21.a2

```python
from numpy import exp
#Declaring Parameters
dV = 27022. #ft/s
pl = 10000. #lbm
g = 32.2 #m/s
staging = [.40*dV,.35*dV,.25*dV]
#system parameters go system letter, Isp, lambda, there is no parameter 4
param = [["a","b","c"],[430.,260.,330.],[.8,.88,.80]]


for k in range(0,3):
    print("staging: "+param[0][k] + param[0][(k+1)%3] + param[0][(k+2)%3])
    massAbove = pl
    values = []
    for i in range(0,3):
        j = (k + i)%3
        print("  stage - " + str(param[0][j]))
        MR = exp(staging[i]/(g*param[1][j]))
        mp = massAbove*((MR-1)/(MR-((MR-1)/param[2][j])))
        print("    mp: " + str(mp))
        mi = (mp/param[2][j])-mp
        print("    mi: " + str(mi))
        print("    total mass of stage: " + str(mp+mi))
        massAbove += mp + mi
        values.append([mp,mi,mp+mi])
    print("final mass for " + param[0][k] + param[0][(k+1)%3] + param[0][(k+2)%3])
    print(str(massAbove))
    print("----------" +"\n")

#Changing permutation to allow for all 6 options
param = [["a","c","b"],[430.,330.,260.],[.8,.80,.88]]

for k in range(0,3):
    print("staging: "+param[0][k] + param[0][(k+1)%3] + param[0][(k+2)%3])
    massAbove = pl
    values = []
    for i in range(0,3):
        j = (k + i)%3
        print("  stage - " + str(param[0][j]))
        MR = exp(staging[i]/(g*param[1][j]))
        mp = massAbove*((MR-1)/(MR-((MR-1)/param[2][j])))
        print("    mp: " + str(mp))
        mi = (mp/param[2][j])-mp
        print("    mi: " + str(mi))
        print("    total mass of stage: " + str(mp+mi))
        massAbove += mp + mi
        values.append([mp,mi,mp+mi])
    print("final mass for " + param[0][k] + param[0][(k+1)%3] + param[0][(k+2)%3])
    print(str(massAbove))
    print("----------" +"\n")
```
